While Loop: Print Until User Enters "stop

```python
import threading


stop_flag = False

def count_numbers():
    count = 1
    while not stop_flag:
        print("Count:", count)
        count += 1

counter_thread = threading.Thread(target=count_numbers)
counter_thread.start()

while True:
    user_input = input("Type 'stop' to end counting: ")
    if user_input.lower() == 'stop':
        stop_flag = True
        break
```

Write a guessing game that asks the user to guess a secret number between 1 and 10.

Give feedback (too high / too low)
Use a while loop

```python
import random

secret_number = random.randint(1, 10)

while True:
    guess = input("Guess the secret number (between 1 and 10): ")


    if not guess.isdigit():
        print("Please enter a valid number.")
        continue

    guess = int(guess)

    if guess < secret_number:
        print("Too low!")
    elif guess > secret_number:
        print("Too high!")
    else:
        print("Correct! You guessed the secret number.")
        break

```

Write a calculator function:
Takes two numbers and an operation (+, -, *, /) Returns the result


```python
def calculator(a, b, operation):
    if operation == '+':
        return a + b
    elif operation == '-':
        return a - b
    elif operation == '*':
        return a * b
    elif operation == '/':
        if b != 0:
            return a / b
        else:
            return "Error: Division by zero"
    else:
        return "Invalid operation"
```

```python
result = calculator(10, 5, '+')
print("Result:", result)  
```

Class and Object Basics
Define a class named Student with the following attributes:
name
age
grade
Create a method display_info() that prints the student's details.
Create three student objects and call the method for each to

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")
```

 Create and Use Student Object

```python
student1 = Student("Maina", 22, "A")
student2 = Student("Amina", 20, "B")
student3 = Student("Brian", 23, "C")

student1.display_info()
student2.display_info()
student3.display_info()
```
